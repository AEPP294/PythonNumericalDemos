#### Linear Regression

Linear regression for prediction.  Here are some key aspects of linear regression:

**Parametric Model**

* the fit model is a simple weighted linear additive model based on all the available features, $x_1,\ldots,x_m$.

* the general form of the multilinear regression model takes the form of $y = \sum_{\alpha = 1}^m b_{\alpha} X_{\alpha} + b_0$

* the specific form of the linear regression model takes the form $y = b_1 x + b_0$

**Least Squares**

* least squares optimization is applied to select the model parameters, $b_1,\ldots,b_m,b_0$ 

* we minize the error over the trainind data $\sum_{i=1}^n (y_i - (\sum_{\alpha = 1}^m b_{\alpha} x_{\alpha} + b_0))^2$

* this could be simplified as the sum of square error over the training data, $\sum_{i=1}^n (\Delta y_i)^2$

**Assumptions**

* **Error-free** - predictor variables are error free, not random variables 
* **Linearity** - response is linear combination of feature(s)
* **Constant Variance** - error in response is constant over predictor(s) value
* **Independence of Error** - error in response are uncorrelated with each other
* **No multicollinearity** - none of the features are redundant with other features 

#### Ridge Regression

With ridge regression we add a hyperparameter, $\lambda$, to our minimization, with a L2 shrinkage penalty term, $\sum_{j=1}^m b_{\alpha}^2$.

\begin{equation}
\sum_{i=1}^n (y_i - (\sum_{\alpha = 1}^m b_{\alpha} x_{\alpha} + b_0))^2 + \lambda \sum_{j=1}^m b_{\alpha}^2
\end{equation}

As a result ridge regression has 2 criteria:

* set the model parameters to minimize the error with training data

* shrink the estimates of the slope parameters towards zero

Note: the intercept is not affected by lambda.

The $\lambda$ is a hyperparameter that controls the degree of fit of the model and may be related to the model variance and bias trade-off.

* for $\lambda \rightarrow 0$ the solution approaches linear regression, there is no bias (relative to a linear model fit), but the variance is high

* as $\lambda$ increases the model variance decreases and the model bias increases

* for $\lambda \rightarrow \infty$ the coefficients approach 0.0 and the model approaches the global mean

#### Lasso Regression

With the lasso we add a hyperparameter, $\lambda$, to our minimization, with a L1 shrinkage penalty term.

\begin{equation}
\sum_{i=1}^n \left(y_i - \left(\sum_{\alpha = 1}^m b_{\alpha} x_{\alpha} + b_0 \right) \right)^2 + \lambda \sum_{j=1}^m |b_{\alpha}|
\end{equation}

As a result the lasso has 2 criteria:

1. set the model parameters to minimize the error with training data

2. shrink the estimates of the slope parameters towards zero. Note: the intercept is not affected by the lambda, $\lambda$, hyperparameter.

Note the only difference between the lasso and ridge regression is:

* for the lasso the shrinkage term is posed as an $\ell_1$ penalty ($\lambda \sum_{\alpha=1}^m |b_{\alpha}|$) 

* for ridge regression the shrinkage term is posed as an $\ell_2$ penalty ($\lambda \sum_{\alpha=1}^m \left(b_{\alpha}\right)^2$).

While both ridge regression and the lasso shrink the model parameters ($b_{\alpha}, \alpha = 1,\ldots,m$) towards zero:

* the lasso parameters reach zero at different rates for each predictor feature as the lambda, $\lambda$, hyperparameter increases. 

* as a result the lasso provides a method for feature ranking and selection!

The lambda, $\lambda$, hyperparameter controls the degree of fit of the model and may be related to the model variance and bias trade-off.

* for $\lambda \rightarrow 0$ the prediction model approaches linear regression, there is lower model bias, but the model variance is higher

* as $\lambda$ increases the model variance decreases and the model bias increases

* for $\lambda \rightarrow \infty$ the coefficients all become 0.0 and the model is the global mean

#### Make a Dataset

Let's make a simple dataset with the following characteristics:

* response feature, $Y$, is a linear combination of predictor features, $X_1$ and $X_2$
* intercept term, $b_0$, is 0.0 and can be neglected, for ease of model parameter space visualization (2D only)

```python
n = 50; seed = 73073
mean_x1 = 10.0; mean_x2 = 10.0; var_x1 = 4.0; var_x2 = 4.0; cov12 = 2.5
db1 = 3.0; db2 = 2.0; eps_std = 0.5
cov = np.array([[var_x1,cov12],[cov12,var_x2]])
np.random.seed(seed = seed)
X = np.random.multivariate_normal(mean = [mean_x1,mean_x2],cov = cov, size = n)
X1, X2 = np.split(X,indices_or_sections = 2,axis=1)
# y = b1*X1 + b2*X2 + np.random.normal(loc = 0.0,scale = eps_std,size = n)
y = np.reshape(db1*X1 + db2*X2,newshape=[n]) + np.reshape(np.random.normal(loc = 0.0,scale = eps_std,size = n),newshape=[n])

sc = plt.scatter(X1,X2,c=y,edgecolor='black',cmap = plt.cm.inferno,vmin=20.0,vmax=60.0)
plt.xlim([5,15]); plt.ylim([5,15]); add_grid(); plt.xlabel('$X_1$'); plt.ylabel('$X_2$'); plt.title('Synthetic Linear Data')
cbar = plt.colorbar(sc, orientation="vertical", ticks=np.linspace(20, 60, 10))
cbar.set_label('$Y$', rotation=270, labelpad=20)

plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=1.0, wspace=0.2, hspace=0.2); plt.show()
```

#### Calculate the Solution Space

Sample of mesh of possible linear regression model parameters and calculate the MSE over all data.

* Note we are not attemption train and test split for model hyperparameter tuning. While we could add a hyper parameter through regularization with LASSO or ridge regression, we just want to visualize this problem.

```python
clam = 0.0 

linear = LinearRegression(fit_intercept = False).fit(X, y)   # fit a linear model
linear_b1, linear_b2 = linear.coef_

ridge = Ridge(alpha=clam,fit_intercept = False).fit(X, y)     # fit a ridge model
ridge_b1, ridge_b2 = ridge.coef_

L = 10
mult_b1 = np.zeros(L); mult_b2 = np.zeros(L); mult_lam = np.zeros(L); 
for i,lam in enumerate(np.logspace(-4,6,L)):
    ridge = Ridge(alpha=lam,fit_intercept = False).fit(X, y)     # fit a ridge model
    mult_b1[i], mult_b2[i] = ridge.coef_; mult_lam[i] = lam
    
vmin = 0.0; vmax = 100.0; #vmax = np.max(MSE_mat)            # set min and max MSE for visualization
norm = colors.Normalize(vmin=vmin, vmax=vmax)
nstep = 200; RSS_mat = np.zeros([nstep,nstep])               # set up the madel parameter mesh
reg_mat = np.zeros([nstep,nstep]); loss_mat = np.zeros([nstep,nstep])  
sb1 = -7.0; eb1 = 7.0; stepb1 = (eb1-sb1)/nstep;         
sb2 = -7.0; eb2 = 7.0; stepb2 = (eb2-sb2)/nstep;
b1_vector = np.arange(sb1, eb1, stepb1); b2_vector = np.arange(eb2, sb2, -1*stepb2)
b1_mat, b2_mat = np.meshgrid(b1_vector, b2_vector)

for ib1, b1 in enumerate(b1_vector):                        # calculate the MSE for all possible model parameters
    for ib2, b2 in enumerate(b2_vector):
        y_hat = np.reshape(b1 * X1 + b2 * X2,newshape=[n])
        RSS_mat[ib2,ib1] =((y - y_hat) ** 2).sum()
        reg_mat[ib2,ib1] = b1*b1 + b2*b2
        loss_mat[ib2,ib1] = RSS_mat[ib2,ib1]+clam*reg_mat[ib2,ib1]

plt.subplot(131)
vmin = np.percentile(RSS_mat.flatten(),q=1); vmax = np.percentile(RSS_mat.flatten(),q=40)
lvmin = math.log10(vmin); lvmax = math.log10(vmax)
im = plt.imshow(RSS_mat,interpolation = None,extent = [sb1,eb1,sb2,eb2],alpha=0.1,vmin=vmin,vmax=vmax,
                cmap = plt.cm.inferno_r,zorder=1)
plt.contour(b1_mat,b2_mat,RSS_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(3,0.3,5),
            colors='black',alpha=0.7,zorder=10)
plt.scatter([linear_b1],[linear_b2],color='black',marker='x',s=30,zorder=100)
plt.title('Ridge RSS L2 Only and Optimal Model Parameters')
plt.xlabel('b1'); plt.ylabel('b2'); add_grid() 

plt.subplot(132)
vmin = np.percentile(reg_mat.flatten(),q=1); vmax = np.percentile(reg_mat.flatten(),q=40)
lvmin = math.log10(vmin); lvmax = math.log10(vmax)
im = plt.imshow(reg_mat,interpolation = None,extent = [sb1,eb1,sb2,eb2],alpha=0.1,vmin=vmin,vmax=vmax,
                cmap = plt.cm.inferno_r,zorder=1)
plt.contour(b1_mat,b2_mat,reg_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(3,0.3,5),
            colors='black',alpha=0.7,zorder=10)
plt.scatter(0.0,0.0,color='black',marker='x',s=30,zorder=100)
plt.title('Ridge Shrinkage L2 Only and Optimal Model Parameters')
plt.xlabel('b1'); plt.ylabel('b2'); add_grid() 

plt.subplot(133)
vmin = np.percentile(loss_mat.flatten(),q=1); vmax = np.percentile(loss_mat.flatten(),q=40)
lvmin = math.log10(vmin); lvmax = math.log10(vmax)
im = plt.imshow(loss_mat,interpolation = None,extent = [sb1,eb1,sb2,eb2],alpha=0.1,vmin=vmin,vmax=vmax,
                cmap = plt.cm.inferno_r,zorder=1)
plt.contour(b1_mat,b2_mat,loss_mat,levels=np.logspace(lvmin,lvmax,5),colors='black',alpha=1.0,
            linewidths=np.linspace(2,0.3,5),zorder=10)
plt.scatter([linear_b1],[linear_b2],color='grey',marker='x',s=30,zorder=100)
plt.plot(mult_b1,mult_b2,color='grey',lw=1,ls='--')
plt.scatter(0.0,0.0,color='grey',marker='x',s=30,zorder=100)

plt.scatter([ridge_b1],[ridge_b2],color='black',marker='x',s=30,zorder=100)
plt.title('Ridge RSS L2 + Shrinkage L2 and Regularized Model Parameters')
plt.xlabel('b1'); plt.ylabel('b2'); add_grid() 

plt.subplots_adjust(left=0.0, bottom=0.0, right=3.0, top=1.2, wspace=0.2, hspace=0.1); plt.show()
```

```python
l = widgets.Text(value='                                       LASSO Regression Regularization Demo, Michael Pyrcz, Professor, The University of Texas at Austin',
        layout=Layout(width='950px', height='30px'))
# P_happening_label = widgets.Text(value='Probability of Happening',layout=Layout(width='50px',height='30px',line-size='0 px'))
clam = widgets.FloatSlider(min=10, max = 3000, value=0, step = 35,description = r'$\lambda$',orientation='horizontal', 
        style = {'description_width':'initial','button_color':'green'},layout=Layout(width='900px',height='40px'),
        continuous_update=False,readout_format='.0f')

ui_summary = widgets.HBox([clam],)
ui_summary3 = widgets.VBox([l,ui_summary],)

def run_plot_summary3(clam):    
        
    linear = LinearRegression(fit_intercept = False).fit(X, y)   # fit a linear model
    linear_b1, linear_b2 = linear.coef_
    
    lasso = Lasso(alpha=clam,fit_intercept = False).fit(X, y)     # fit a ridge model
    lasso_b1, lasso_b2 = lasso.coef_
    
    L = 10
    mult_b1 = np.zeros(L); mult_b2 = np.zeros(L); mult_lam = np.zeros(L); 
    for i,lam in enumerate(np.linspace(10,600,L)):
        lasso = Lasso(alpha=lam,fit_intercept = False).fit(X, y)     # fit a ridge model
        mult_b1[i], mult_b2[i] = lasso.coef_; mult_lam[i] = lam
        
    vmin = 0.0; vmax = 100.0; #vmax = np.max(MSE_mat)            # set min and max MSE for visualization
    norm = colors.Normalize(vmin=vmin, vmax=vmax)
    nstep = 200; RSS_mat = np.zeros([nstep,nstep])               # set up the madel parameter mesh
    reg_mat = np.zeros([nstep,nstep]); loss_mat = np.zeros([nstep,nstep])  
    sb1 = -7.0; eb1 = 7.0; stepb1 = (eb1-sb1)/nstep;         
    sb2 = -7.0; eb2 = 7.0; stepb2 = (eb2-sb2)/nstep;
    b1_vector = np.arange(sb1, eb1, stepb1); b2_vector = np.arange(eb2, sb2, -1*stepb2)
    b1_mat, b2_mat = np.meshgrid(b1_vector, b2_vector)
    
    for ib1, b1 in enumerate(b1_vector):                        # calculate the MSE for all possible model parameters
        for ib2, b2 in enumerate(b2_vector):
            y_hat = np.reshape(b1 * X1 + b2 * X2,newshape=[n])
            RSS_mat[ib2,ib1] =((y - y_hat) ** 2).sum()
            reg_mat[ib2,ib1] = abs(b1) + abs(b2)
            loss_mat[ib2,ib1] = (1/len(y))*RSS_mat[ib2,ib1] + 2*clam*reg_mat[ib2,ib1]
    
    plt.subplot(131)
    vmin = np.percentile(RSS_mat.flatten(),q=1); vmax = np.percentile(RSS_mat.flatten(),q=40)
    lvmin = math.log10(vmin); lvmax = math.log10(vmax)
    im = plt.imshow(RSS_mat,interpolation = None,extent = [sb1,eb1,sb2,eb2],alpha=0.1,vmin=vmin,vmax=vmax,
                    cmap = plt.cm.inferno_r,zorder=1)
    plt.contour(b1_mat,b2_mat,RSS_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(3,0.3,5),
                colors='black',alpha=0.7,zorder=10)
    plt.scatter([linear_b1],[linear_b2],color='black',marker='x',s=30,zorder=100)
    plt.title('Ridge RSS L2 Only and Optimal Model Parameters')
    plt.xlabel('b1'); plt.ylabel('b2'); add_grid() 
    
    plt.subplot(132)
    vmin = np.percentile(reg_mat.flatten(),q=1); vmax = np.percentile(reg_mat.flatten(),q=40)
    lvmin = math.log10(vmin); lvmax = math.log10(vmax)
    im = plt.imshow(reg_mat,interpolation = None,extent = [sb1,eb1,sb2,eb2],alpha=0.1,vmin=vmin,vmax=vmax,
                    cmap = plt.cm.inferno_r,zorder=1)
    plt.contour(b1_mat,b2_mat,reg_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(3,0.3,5),
                colors='black',alpha=0.7,zorder=10)
    plt.scatter(0.0,0.0,color='black',marker='x',s=30,zorder=100)
    plt.title('Ridge Shrinkage L2 Only and Optimal Model Parameters')
    plt.xlabel('b1'); plt.ylabel('b2'); add_grid() 
    
    plt.subplot(133)
    vmin = np.percentile(loss_mat.flatten(),q=1); vmax = np.percentile(loss_mat.flatten(),q=40)
    lvmin = math.log10(vmin); lvmax = math.log10(vmax)
    im = plt.imshow(loss_mat,interpolation = None,extent = [sb1,eb1,sb2,eb2],alpha=0.1,vmin=vmin,vmax=vmax,
                    cmap = plt.cm.inferno_r,zorder=1)
    plt.contour(b1_mat,b2_mat,loss_mat,levels=np.logspace(lvmin,lvmax,5),colors='black',alpha=1.0,
                linewidths=np.linspace(2,0.3,5),zorder=10)
    plt.scatter([linear_b1],[linear_b2],color='grey',marker='x',s=30,zorder=100)
    plt.plot(mult_b1,mult_b2,color='grey',lw=1,ls='--')
    plt.scatter(0.0,0.0,color='grey',marker='x',s=30,zorder=100)
    
    plt.scatter([lasso_b1],[lasso_b2],color='black',marker='x',s=30,zorder=100)
    plt.title('Ridge RSS L2 + Shrinkage L2 and Regularized Model Parameters')
    plt.xlabel('b1'); plt.ylabel('b2'); add_grid() 
    
    plt.subplots_adjust(left=0.0, bottom=0.0, right=3.0, top=1.2, wspace=0.2, hspace=0.1); plt.show()
    
interactive_plot_summary3 = widgets.interactive_output(run_plot_summary3, {'clam':clam,})
interactive_plot_summary3.clear_output(wait = True)  
```

#### Interactive Dashboard - Ridge vs. LASSO OLS vs. Regularization

Let's start by visualizing the change in OLS vs. regularization components of the loss function for ridge regression and LASSO regression as we change the hyperparameter, $\lambda$.

* we use on 2 predictor feature and assume the intercept is 0.0 ($b_0 = 0.0$) so we can conveniently visualize the entire model parameter space in 2D.

```python
l = widgets.Text(value='                                       Ridge L2 and LASSO L1 Regression Regularization Demo, Michael Pyrcz, Professor, The University of Texas at Austin',
        layout=Layout(width='950px', height='30px'))
# P_happening_label = widgets.Text(value='Probability of Happening',layout=Layout(width='50px',height='30px',line-size='0 px'))
slam = widgets.IntSlider(min=1, max = 20, value=1, step = 1.0,description = r'$\lambda$ level',orientation='horizontal', 
        style = {'description_width':'initial','button_color':'green'},layout=Layout(width='900px',height='40px'),
        continuous_update=False,readout_format='.0f')

ui_summary = widgets.HBox([slam],)
ui_summary2 = widgets.VBox([l,ui_summary],)

def run_plot_summary2(slam):   
    ridge_lam_mat = np.logspace(1,6,20); lasso_lam_mat = np.linspace(1,600,20)
    ridge_lam = ridge_lam_mat[slam-1]; lasso_lam = lasso_lam_mat[slam-1];
    
    # calculate paths over multiple hyperparameters
    L = 100
    ridge_mult_b1 = np.zeros(L); ridge_mult_b2 = np.zeros(L); ridge_mult_lam = np.zeros(L) 
    lasso_mult_b1 = np.zeros(L); lasso_mult_b2 = np.zeros(L); lasso_mult_lam = np.zeros(L)
    for i,lam in enumerate(np.logspace(-4,6,L)):
        ridge = Ridge(alpha=lam,fit_intercept = False).fit(X, y)     # fit a ridge model
        ridge_mult_b1[i], ridge_mult_b2[i] = ridge.coef_; ridge_mult_lam[i] = lam
        
    for i,lam in enumerate(np.logspace(0,3,L)):
        lasso = Lasso(alpha=lam,fit_intercept = False).fit(X, y)     # fit a LASSO model
        lasso_mult_b1[i], lasso_mult_b2[i] = lasso.coef_; lasso_mult_lam[i] = lam
        
    # fit linear, Ridge and LASSO models
    linear = LinearRegression(fit_intercept = False).fit(X, y)   # fit a linear model
    linear_b1, linear_b2 = linear.coef_
    
    ridge = Ridge(alpha=ridge_lam,fit_intercept = False).fit(X, y)     # fit a ridge model
    ridge_b1, ridge_b2 = ridge.coef_
    
    lasso = Lasso(alpha=lasso_lam,fit_intercept = False).fit(X, y)     # fit a ridge model
    lasso_b1, lasso_b2 = lasso.coef_    
    
    vmin = 0.0; vmax = 100.0; #vmax = np.max(MSE_mat)            # set min and max MSE for visualization
    norm = colors.Normalize(vmin=vmin, vmax=vmax)
    nstep = 200; RSS_mat = np.zeros([nstep,nstep])               # set up the madel parameter mesh
    ridge_reg_mat = np.zeros([nstep,nstep]); ridge_loss_mat = np.zeros([nstep,nstep])  
    lasso_reg_mat = np.zeros([nstep,nstep]); lasso_loss_mat = np.zeros([nstep,nstep])  
    sb1 = -7.0; eb1 = 7.0; stepb1 = (eb1-sb1)/nstep;         
    sb2 = -7.0; eb2 = 7.0; stepb2 = (eb2-sb2)/nstep;
    b1_vector = np.arange(sb1, eb1, stepb1); b2_vector = np.arange(eb2, sb2, -1*stepb2)
    b1_mat, b2_mat = np.meshgrid(b1_vector, b2_vector)
    
    for ib1, b1 in enumerate(b1_vector):                        # calculate the MSE for all possible model parameters
        for ib2, b2 in enumerate(b2_vector):
            y_hat = np.reshape(b1 * X1 + b2 * X2,newshape=[n])
            RSS_mat[ib2,ib1] =((y - y_hat) ** 2).sum()
            ridge_reg_mat[ib2,ib1] = b1*b1 + b2*b2
            ridge_loss_mat[ib2,ib1] = RSS_mat[ib2,ib1]+ridge_lam*ridge_reg_mat[ib2,ib1]
            
            lasso_reg_mat[ib2,ib1] = abs(b1) + abs(b2)
            lasso_loss_mat[ib2,ib1] = RSS_mat[ib2,ib1]+lasso_lam*lasso_reg_mat[ib2,ib1]
    
    ### find the solution
    iridge_b1 = (np.absolute(b1_vector-ridge_b1)).argmin(); iridge_b2 = (np.absolute(b2_vector-ridge_b2)).argmin()
    ridge_RSS = RSS_mat[iridge_b2,iridge_b1]; ridge_reg = ridge_reg_mat[iridge_b2,iridge_b1]
    
    ilasso_b1 = (np.absolute(b1_vector-lasso_b1)).argmin(); ilasso_b2 = (np.absolute(b2_vector-lasso_b2)).argmin()
    lasso_RSS = RSS_mat[ilasso_b2,ilasso_b1]; lasso_reg = lasso_reg_mat[ilasso_b2,ilasso_b1]
    
    plt.subplot(121)
    
    vmin = np.percentile(RSS_mat.flatten(),q=1); vmax = np.percentile(RSS_mat.flatten(),q=20)
    lvmin = math.log10(vmin); lvmax = math.log10(vmax)
    plt.contour(b1_mat,b2_mat,RSS_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(0.8,0.2,5),
                colors='darkred',linestyles='dotted',alpha=0.4,zorder=10)
    vmin = np.percentile(ridge_reg_mat.flatten(),q=1); vmax = np.percentile(ridge_reg_mat.flatten(),q=20)
    lvmin = math.log10(vmin); lvmax = math.log10(vmax)
    plt.contour(b1_mat,b2_mat,ridge_reg_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(0.8,0.2,5),
                colors='#cc5500',alpha=0.4,zorder=10)
    
    plt.contour(b1_mat,b2_mat,RSS_mat,levels = [ridge_RSS],linewidths=2.0,
                 colors='darkred',alpha=0.7,zorder=10)
    plt.scatter([linear_b1],[linear_b2],color='darkred',marker='x',s=50,zorder=100)
    
    plt.contour(b1_mat,b2_mat,ridge_reg_mat,levels = [ridge_reg],linewidths=2.0,
                 colors='#cc5500',alpha=0.7,zorder=10)
    plt.scatter([0.0],[0.0],color='#cc5500',marker='x',s=50,zorder=100)
    
    plt.scatter([ridge_b1],[ridge_b2],color='blue',edgecolor='black',marker='o',s=30,zorder=1000)
    plt.plot(ridge_mult_b1,ridge_mult_b2,color='black',lw=1)
    
    plt.annotate('Global \n Mean',(0.8,-0.5),ha='center',color = '#cc5500')
    plt.annotate('Ordinary Least \n Square',(linear_b1 + 1.6,linear_b2 - 0.5), ha='center',color='darkred')
    
    plt.plot([ridge_b1,ridge_b1],[sb1,ridge_b2],color='blue',lw=1.0,ls='--')
    plt.annotate(r'$b_1$ = ' + str(np.round(ridge_b1,1)),(ridge_b1+0.2,-6.0),ha='center',color = 'blue',rotation=270)
    
    plt.plot([sb1,ridge_b1],[ridge_b2,ridge_b2],color='blue',lw=1.0,ls='--')
    plt.annotate(r'$b_2$ = ' + str(np.round(ridge_b2,1)),(-6.0,ridge_b2+0.2),ha='center',color = 'blue')
    
    plt.title(r'Ridge Regression RSS L2 and Regularization L2 for $\lambda = $' + str(np.round(ridge_lam)))
    plt.xlabel(r'$b_1$ Model Parameter'); plt.ylabel(r'$b_2$ Model Parameter'); add_grid() 
    plt.xlim([sb1,eb1]); plt.ylim([sb2,eb2])
    
    plt.subplot(122)
    
    vmin = np.percentile(RSS_mat.flatten(),q=1); vmax = np.percentile(RSS_mat.flatten(),q=20)
    lvmin = math.log10(vmin); lvmax = math.log10(vmax)
    plt.contour(b1_mat,b2_mat,RSS_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(0.8,0.2,5),
                colors='darkred',linestyles='dotted',alpha=0.4,zorder=10)
    vmin = np.percentile(lasso_reg_mat.flatten(),q=1); vmax = np.percentile(lasso_reg_mat.flatten(),q=20)
    lvmin = math.log10(vmin); lvmax = math.log10(vmax)
    plt.contour(b1_mat,b2_mat,lasso_reg_mat,levels = np.logspace(lvmin,lvmax,5),linewidths=np.linspace(0.8,0.2,5),
                colors='#cc5500',alpha=0.4,zorder=10)
    
    plt.contour(b1_mat,b2_mat,RSS_mat,levels = [lasso_RSS],linewidths=2.0,
                 colors='darkred',alpha=0.7,zorder=10)
    plt.scatter([linear_b1],[linear_b2],color='darkred',marker='x',s=50,zorder=100)
    
    plt.contour(b1_mat,b2_mat,lasso_reg_mat,levels = [lasso_reg],linewidths=2.0,
                 colors='#cc5500',alpha=0.7,zorder=10)
    plt.scatter([0.0],[0.0],color='#cc5500',marker='x',s=50,zorder=100)
    
    plt.scatter([lasso_b1],[lasso_b2],color='blue',edgecolor='black',marker='o',s=30,zorder=1000)
    plt.plot(lasso_mult_b1,lasso_mult_b2,color='black',lw=1)
    
    plt.annotate('Global \n Mean',(0.8,-0.5),ha='center',color = '#cc5500')
    plt.annotate('Ordinary Least \n Square',(linear_b1 + 1.6,linear_b2 - 0.5), ha='center',color='darkred')
    
    plt.plot([lasso_b1,lasso_b1],[sb1,lasso_b2],color='blue',lw=1.0,ls='--')
    plt.annotate(r'$b_1$ = ' + str(np.round(lasso_b1,1)),(lasso_b1+0.2,-6.0),ha='center',color = 'blue',rotation=270)
    
    plt.plot([sb1,lasso_b1],[lasso_b2,lasso_b2],color='blue',lw=1.0,ls='--')
    plt.annotate(r'$b_2$ = ' + str(np.round(lasso_b2,1)),(-6.0,lasso_b2+0.2),ha='center',color = 'blue')
    
    plt.title(r'LASSO Regression RSS L2 and Regularization L1 for $\lambda = $' + str(np.round(lasso_lam,1)))
    plt.xlabel(r'$b_1$ Model Parameter'); plt.ylabel(r'$b_2$ Model Parameter'); add_grid() 
    plt.xlim([sb1,eb1]); plt.ylim([sb2,eb2])
    
    plt.subplots_adjust(left=0.0, bottom=0.0, right=2.0, top=1.2, wspace=0.2, hspace=0.1); plt.show()
    
interactive_plot_summary2 = widgets.interactive_output(run_plot_summary2, {'slam':slam,})
interactive_plot_summary.clear_output(wait = True)  
```
